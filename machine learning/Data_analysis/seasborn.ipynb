```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
# Basic plotting with seaborn
tips = sns.load_dataset('tips')
tips
```

```python
# Create a scatter plot
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
plt.title("Toalt bill vs tips scatter plot")
plt.show
```

```python
# Line Plot 
sns.lineplot(x = 'size', y = 'total_bill', data = tips)
plt.title("size vs total bill lineplot")
plt.plot()
```

```python
# Categorical plot
# bar plot
sns.barplot(x = 'day', y = 'total_bill', data = tips) 
plt.title("Day vs total bill barplot")
plt.show()
```

```python
# Box plot
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.title("Box plot")
plt.show()
```

```python
# Violine plot
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
plt.title("Violin plot")
plt.show()
```

```python
# Histogram
sns.histplot(tips['total_bill'],bins = 10, kde = True)  # kde = kernel density estimate
```

```python
# KDE Plot
sns.kdeplot(tips['total_bill'], fill = True)
```

```python
# Pair plot
sns.pairplot(tips)
```

```python
# Heatmap
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot = True, cmap = 'coolwarm')
```

```python
# Male vs female tip 
sns.barplot(x = 'sex',y = 'tip', data = tips,color='skyblue',edgecolor='black') 
plt.xlabel('gender', fontsize=12)
plt.ylabel('tip', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.title('Tips vs genders')
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex")
plt.title("Tips by Gender")
plt.show()
```

```python
import pandas as pd
sales_data = pd.read_csv('sales_data.csv',encoding='latin1')
sales_data.head()
```

```python
sales_data.info()
```

```python
# Plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x = 'PRODUCTLINE',y = 'SALES', data = sales_data, estimator = sum)
plt.title('Product vs sales')
plt.xlabel('Product')
plt.ylabel('sales')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x = 'COUNTRY',y = 'SALES', data = sales_data, estimator = sum)
plt.title('country vs sales')
plt.xlabel('country')
plt.ylabel('sales')
plt.show()
```
